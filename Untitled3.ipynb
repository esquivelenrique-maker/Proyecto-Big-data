```python
# Regresión: Medical Cost Personal Datasets
df_enrique_regresion = pd.read_csv('insurance.csv')

# Clasificación: Heart Disease UCI
df_enrique_clasificacion = pd.read_csv('heart_disease_uci.csv')
```

```python
# Código para el EDA de Clasificación
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribución por Sexo (sex)
sns.countplot(x='sex', hue='target', data=df_enrique_clasificacion, ax=axes[0], palette='viridis')
axes[0].set_title('Conteo de Enfermedad Cardíaca por Sexo')
axes[0].set_xlabel('Sexo (0=Mujer, 1=Hombre)')

# Distribución por Tipo de Dolor de Pecho (cp)
sns.countplot(x='cp', hue='target', data=df_enrique_clasificacion, ax=axes[1], palette='viridis')
axes[1].set_title('Conteo de Enfermedad Cardíaca por Tipo de Dolor de Pecho (cp)')

# Distribución por Angina inducida por ejercicio (exang)
sns.countplot(x='exang', hue='target', data=df_enrique_clasificacion, ax=axes[2], palette='viridis')
axes[2].set_title('Conteo de Enfermedad Cardíaca por Angina inducida por ejercicio (exang)')

plt.tight_layout()
plt.show() #
```

```python
# Separar la variable predictora (X) y la variable objetivo (y)
X_reg = df_enrique_regresion.drop('charges', axis=1)
y_reg = df_enrique_regresion['charges']

# Identificar columnas
categorical_features = ['sex', 'smoker', 'region']
numerical_features = ['age', 'bmi', 'children']

# Crear el Preprocesador usando ColumnTransformer
preprocessor_reg = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('scaler', StandardScaler(), numerical_features)
    ],
    remainder='passthrough'
)

# Segmentación de Datos (Requisito: Usar X_train_E, y_test_E, etc.)
X_train_E_reg, X_test_E_reg, y_train_E_reg, y_test_E_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

```python
# Crear el Pipeline de Regresión
regresor_esquivel = Pipeline(steps=[
    ('preprocessor', preprocessor_reg),
    ('regressor', LinearRegression())
])

# Entrenamiento del Modelo
regresor_esquivel.fit(X_train_E_reg, y_train_E_reg)

# Predicciones
y_pred_E_reg = regresor_esquivel.predict(X_test_E_reg)

# Evaluación del Modelo
r2 = r2_score(y_test_E_reg, y_pred_E_reg)

# Calcular RMSE aplicando la raíz cuadrada al MSE
mse = mean_squared_error(y_test_E_reg, y_pred_E_reg)
rmse = np.sqrt(mse)

print("\n--- Evaluación del Modelo de Regresión (regresor_esquivel) ---")
print(f"R2 Score: {r2:.4f}")
print(f"RMSE (Error Cuadrático Medio): ${rmse:.2f}")
```
